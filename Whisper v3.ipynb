```python
def convert_text_to_srt(text, output_file):
    lines = text.strip().split('\n')

    # Generate subtitle timing
    start_time = 0  # Initial start time in milliseconds
    interval = chunk_unit  # Time interval between subtitles in milliseconds
    end_time = chunk_unit  # Subtitle duration in milliseconds

    subtitles = []
    for i, line in enumerate(lines):
        subtitle_text = line.strip()
        start = format_time(start_time)
        end = format_time(min(end_time, total_duration_seconds_millseconds))  # Use the minimum of end_time or total_duration as the actual ending time
        subtitle = f"{i + 1}\n{start} --> {end}\n{subtitle_text}\n"
        subtitles.append(subtitle)

        start_time += interval
        end_time += interval

    # Write subtitles to output file with 'utf-8' encoding
    with open(output_file, 'w', encoding='utf-8') as file:
        file.writelines(subtitles)


def format_time(milliseconds):
    total_seconds = milliseconds // 1000
    minutes, seconds = divmod(total_seconds, 60)
    hours, minutes = divmod(minutes, 60)
    formatted_time = f"{hours:02d}:{minutes:02d}:{seconds:02d},{milliseconds % 1000:03d}"
    return formatted_time


# Example usage
text = ' '.join(translate_subtitle)

output_file = "output/SRT_translate_output.srt"

convert_text_to_srt(text, output_file)
```
